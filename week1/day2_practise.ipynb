```python
# ─────────────────────────────────────────────
# EXERCISE 1: Grade Classifier
# ─────────────────────────────────────────────
# Given a list of student scores, print each score with its grade:
# >= 90 → "A", >= 80 → "B", >= 70 → "C", >= 60 → "D", else → "F"
# Also print how many students passed (grade != F)

scores = [92, 45, 78, 85, 61, 55, 90, 73, 48, 88]

# YOUR CODE HERE:
for i in scores:
    if i >= 90:
        print("A")
    elif i >= 80:
        print("B")
    elif i>= 70:
        print("C") 
    elif i>= 60:
        print("D")
    else:
        print("F")
```

```python
# ─────────────────────────────────────────────
# EXERCISE 2: FizzBuzz (Data Edition)
# ─────────────────────────────────────────────
# For numbers 1 to 30:
#   - If divisible by 3 AND 5 → print "DataAnalyst"
#   - If divisible by 3 only  → print "Data"
#   - If divisible by 5 only  → print "Analyst"
#   - Otherwise               → print the number

# YOUR CODE HERE:

for i in range(31):
    if i % 3 == 0 and i % 5 == 0:
        print("DataAnalyst")
    elif i % 3 == 0:
        print("Data")
    elif i % 5 == 0:
        print("Analyst")
    else:
        print(i)
```

```python
# ─────────────────────────────────────────────
# EXERCISE 3: Sales Summary with Loop
# ─────────────────────────────────────────────
# Given this sales data, using a loop calculate:
# 1. Total revenue
# 2. Average revenue
# 3. Highest sale (and which product)
# 4. Lowest sale (and which product)
# 5. Number of products with revenue > 50000

from altair import value
from numpy import average


sales = {
    "Laptop":   120000,
    "Phone":    45000,
    "Tablet":   78000,
    "Monitor":  22000,
    "Keyboard": 15000,
    "Headphones": 61000,
}


# YOUR CODE HERE:

total_revenue = 0
average_revenue = 0
best_product = ""
best_amount = 0
worst_product = ""
worst_amount = float("inf")
count = 0



for amount in sales.values():
    total_revenue = total_revenue + amount
    average_revenue = total_revenue/6
for product, amount in sales.items():
    if amount > best_amount:
        best_amount = amount
        best_product = product
for product, amount in sales.items():
    if amount < worst_amount:
        worst_amount = amount
        worst_product = product

for amount in sales.values():
    if amount > 50000:
        count += 1



print(total_revenue)
print(average_revenue)
print(best_product, best_amount)
print(worst_product, worst_amount)
print(count)
```
